The program below takes client data that was being stored in an Excel file and restructures and formats the data in a fashion that will ultimately be uploaded into a database.    

```python
import pandas as pd
import numpy as np
```

```python
#The dataframes(enrollment and personnel) are there for reference as they provide field names and order of them.
#They each represent a table in the database the data will be uploaded into 

enrollment = pd.read_csv("df_enrollment_info.csv") 
personnel = pd.read_csv("df_personnel_info.csv")
```

```python
#Read in client data. Two dataframes are created as each will be uploaded into its respected table in the database
workability_e = pd.read_csv("HS_Workability_Program.csv")
workability_p = pd.read_csv("HS_Workability_Program.csv")
```

Enrollment Dataframe

```python
workability_e.rename(columns={"Start Date":"start_date", "Store":"site", "School District":"referral_type",
                           "Status":"enrollment_status","End Date":"actual_close_date","Last Name":
                              "last_name"}, inplace = True) #Rename columns to align in database
```

```python
#Inserted new columns that are present in database
workability_e.insert(0, column = ("wds_id"),value=str)
workability_e.insert(1, column = ("enrollment_id"),value=str)
workability_e.insert(2, column = ("department_number"),value=str)
workability_e.insert(3, column = ("gw_or_offsite"),value=str)
workability_e.insert(5, column = ("referral_date"),value=str)
workability_e.insert(7, column = ("projected_end_date"),value=str)
workability_e.insert(10, column = ("hours"),value=str)
workability_e.insert(11, column = ("case_number"),value=str)
workability_e.insert(13, column = ("skills_trainer"),value=str)
workability_e.insert(14, column = ("case_worker"),value=str)
workability_e.insert(15, column = ("Homeless"),value=str)
workability_e.insert(16, column = ("employed"),value=str)
workability_e.insert(17, column = ("date_of_hire"),value=str)
workability_e.insert(18, column = ("employer"),value=str)
workability_e.insert(19, column = ("hours_per_week"),value=str)
workability_e.insert(20, column = ("wage"),value=str)
workability_e.insert(21, column = ("job_title"),value=str)
workability_e.insert(22, column = ("reason_not_completed"),value=str)
workability_e.insert(23, column = ("naics_code"),value=str)
workability_e.insert(24, column = ("initial_eval"),value=str)
workability_e.insert(25, column = ("final_eval"),value=str)
workability_e.insert(26, column = ("wages_paid"),value=str)
workability_e.insert(27, column = ("digital_literacy_enrollment"),value=str)
workability_e.insert(28, column = ("digital_certificate"),value=str)

```

```python
workability_e.drop(columns = (["First Name", "Address", "City", "Phone Number", 
                               "Zip Code","Position", "Pay", "case_worker", "hours", "Date of Position"]), inplace= True) #dropped columns no longer needed
```

```python

workability_e.rename(columns = {"Total Assigned Hours":"hours", "Transition Specialist": "case_worker"}, inplace = True) #Rename several of the fields
```

```python
#Reindex fields and put them in proper order as table in database
workability_e = workability_e.reindex(columns=['wds_id',"last_name", 'enrollment_id', 'department_number', 'gw_or_offsite',
                               'referral_type', 'referral_date', 'start_date', 'projected_end_date',
                               'actual_close_date','enrollment_status', 'hours', 'case_number', "site", "skills_trainer",
                              "case_worker", "Homeless", "employed", "date_of_hire", "employer", "hours_per_week",
                              "wage", "job_title", "reason_not_completed", "naics_code", "initial_eval", "final_eval",
                              "wages_paid", "digital_literacy_enrollment", "digital_certificate"])
```

```python
#Remove extra spaces in field headers
for i in range(0,30):
    workability_e.iloc[:,i] = workability_e.iloc[:,i].str.strip()
```

```python
#Insert values for specific fields
workability_e["actual_close_date"].iloc[145] = np.nan
workability_e["actual_close_date"].iloc[76] = np.nan 
workability_e["department_number"] = "Workability"
workability_e["enrollment_status"] = "Completed"
```

```python
#Update fields to correct datatypes 
workability_e["referral_date"] = pd.to_datetime(workability_e["referral_date"])
workability_e["start_date"] = pd.to_datetime(workability_e["start_date"])
workability_e["projected_end_date"] = pd.to_datetime(workability_e["projected_end_date"])
workability_e["actual_close_date"] = pd.to_datetime(workability_e["actual_close_date"])
workability_e["date_of_hire"] = pd.to_datetime(workability_e["date_of_hire"])
```

Personnel Dataframe

```python
#Rename fields to match personnel table in database
workability_p.rename(columns = {"Last Name":"last_name", "First Name": "first_name", "Address":"address",
                                "City":"city", "Phone Number": "phone", "Zip Code":"zip"}, inplace=True)
```

```python
#Drop fields not present in table in database
workability_p.drop(columns = (["Start Date", "Store", "School District", "Pay", "Status", "Date of Position",
                              "Position", "End Date", "Transition Specialist", "Total Assigned Hours"]), 
                   axis = 1, inplace = True)
```

```python
#Insert new fields to conform to table in database
workability_p.insert(0, column = ("wds_id"),value=str)
workability_p.insert(1, column = ("paycom_id"),value=str)
workability_p.insert(4, column = ("ssn"),value=str)
workability_p.insert(5, column = ("dob"),value=str)
workability_p.insert(6, column = ("email"),value=str)
workability_p.insert(7, column = ("ethnicity"),value=str)
workability_p.insert(8, column = ("gender"),value=str)
workability_p.insert(10, column = ("case_number"),value=str)
workability_p.insert(13, column = ("state"),value=str)
workability_p.insert(15, column = ("lanuage"),value=str)
```

```python
#Remove extra spaces in field headers
for i in range(0,16):
    workability_p.iloc[:,i] = workability_p.iloc[:,i].str.strip()
```

```python
#Send each dataframe to csv file that will be uploaded into company database
workability_e.to_csv("Final_Enrollment.csv")
workability_p.to_csv("Final_Personnel.csv")
```
